```python
df.columns = df.columns.str.strip()
```

```python
# Create new columns for Total, Maximum, Minimum, and Average marks
df['Total'] = df[['math score', 'reading score', 'writing score']].sum(axis=1)
df['Maximum'] = df[['math score', 'reading score', 'writing score']].max(axis=1)
df['Minimum'] = df[['math score', 'reading score', 'writing score']].min(axis=1)
df['Average'] = df[['math score', 'reading score', 'writing score']].mean(axis=1)


```

```python
# Sort the DataFrame by the 'Average' column in descending order
df_sorted = df.sort_values(by='Average', ascending=False)


```

```python
# Change 'female' to 'F' and 'male' to 'M' in the 'gender' column
df_sorted['gender'] = df['gender'].replace({'female': 'F', 'male': 'M'})

# Remove the word 'group' from the 'race/ethnicity' column
df_sorted['race/ethnicity'] = df['race/ethnicity'].str.replace('group ', '')

```

```python
df_sorted.head()
```

```python
# Select the first 100 rows
subset = df_sorted.head(100)

# Count the unique values in the 'gender' column
gender_count = subset['gender'].value_counts()

# Count the unique values in the 'race/ethnicity' column
race_count = subset['race/ethnicity'].value_counts()


```

```python
# Display the counts
print("Gender Count (First 100 Entries):")
print(gender_count)

print("\nRace/Ethnicity Count (First 100 Entries):")
print(race_count)
```
